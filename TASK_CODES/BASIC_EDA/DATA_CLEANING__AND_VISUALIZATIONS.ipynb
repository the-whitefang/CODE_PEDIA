```python
df['PRODUCTLINE'].head(35)
```

```python
most_sold = df.groupby('PRODUCTCODE')['QUANTITYORDERED'].sum().sort_values(ascending=False)
most_sold
```

```python
most_sold.head()
```

```python
top_5_products = most_sold.head(5).reset_index()
top_5_products
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='PRODUCTCODE', y='QUANTITYORDERED', data=top_5_products, palette='viridis')
plt.title('Top 5 Most Sold Products by Quantity')
plt.xlabel('Product Code')
plt.ylabel('Total Quantity Ordered')
plt.tight_layout()
plt.show()
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sales_by_productline = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values()
sns.barplot(x=sales_by_productline.values, y=sales_by_productline.index, palette='Blues_d')
plt.title('Total Sales by Product Line')
plt.xlabel('Total Sales')
plt.ylabel('Product Line')
plt.tight_layout()
plt.show()
```

```python
monthly_sales = df.set_index('ORDERDATE').resample('M')['SALES'].sum()

plt.figure(figsize=(12, 6))
monthly_sales.plot(marker='o', color='teal')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='DEALSIZE', y='SALES', data=df, palette='Set2')
plt.title('Sales Distribution by Deal Size')
plt.xlabel('Deal Size')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
top_customers = df.groupby('CUSTOMERNAME')['SALES'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_customers.values, y=top_customers.index, palette='viridis')
plt.title('Top 10 Customers by Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Customer Name')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
avg_price_by_product = df.groupby('PRODUCTLINE')['PRICEEACH'].mean().sort_values()
sns.barplot(x=avg_price_by_product.values, y=avg_price_by_product.index, palette='coolwarm')
plt.title('Average Price per Product Line')
plt.xlabel('Average Price Each')
plt.ylabel('Product Line')
plt.tight_layout()
plt.show()
```

```python
country_sales = df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=country_sales.values, y=country_sales.index, palette='crest')
plt.title('Top 10 Countries by Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
